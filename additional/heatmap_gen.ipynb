Create Heatmap based off of tsv file with additional annotations


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the input file
tally_vals = pd.read_csv("/Users/rwura/bioinfo/cross_lab/rev_tax_pipeline/test/data/redone_matrix.tsv", sep='\t')

df = tally_vals[['Barcode','Scientific_Name','Normalized_tally']]

df = df[df.Normalized_tally > 3]
df = df.pivot(index='Barcode',columns='Scientific_Name',values='Normalized_tally')
plt.figure(figsize=(8, 12))
sns.heatmap(df, cmap="viridis", annot=True, fmt='.0f', linewidths=.5)

# Customize the plot
plt.xticks(rotation=90, fontsize=8)
plt.yticks(rotation=0, fontsize=8)
plt.xlabel('')
plt.ylabel('Barcode')
plt.title('Abundance of viruses with tally counts over 10')


# Show the plot
plt.tight_layout()
plt.savefig('maine_species_data.png')
```
